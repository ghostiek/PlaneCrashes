```python
#Operators that crash the most total
sns.set_style("whitegrid")
sns.set(font_scale=2)

plt.figure(figsize=(20,15))
ax = sns.countplot(x="Operator", data=fixed_df, order=fixed_df.Operator.value_counts().iloc[:10].index)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_title('Top 10 operator')
ax.set_ylabel('Count')
ax.set_xlabel('Operator')
```

```python
crashes_per_year = fixed_df.groupby([fixed_df.index.year, "Operator"]).sum()
last_ten_years = crashes_per_year.loc[(slice('2000','2009'),), :]
last_ten_years
```
